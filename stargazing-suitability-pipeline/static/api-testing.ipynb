```python
import requests
import json

API_KEY = ""
LAT, LON = 31.5204, 74.3587

url = f"https://api.openweathermap.org/data/3.0/onecall?lat={LAT}&lon={LON}&exclude=minutely,daily,alerts&units=metric&appid={API_KEY}"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    city = "Lahore"
    
    # Use current weather (or hourly[0] if you prefer)
    current = data["current"]
    clouds = current["clouds"]
    humidity = current["humidity"]
    visibility = current["visibility"]  # in meters

    # Stargazing score calculation
    score = (1 - clouds / 100) * (1 - humidity / 100) * (visibility / 10000)
    score = round(score, 3)

    # Package message
    message = {
        "city": city,
        "timestamp": current["dt"],
        "clouds": clouds,
        "humidity": humidity,
        "visibility": visibility,
        "score": score,
        "summary": current["weather"][0]["description"]
    }

    print("🌌 Stargazing Score:", message)
else:
    print("❌ API Error:", response.status_code)
```
