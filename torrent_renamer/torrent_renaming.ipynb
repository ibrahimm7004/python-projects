```python
import openai
import os
import pandas as pd
openai.api_key = ''
```

```python
path = r'C:\Users\hp\Downloads\torrents'
all_entries = os.listdir(path)
df = pd.DataFrame(columns=['Old Name', 'New Name'])
```

```python
def reformat_names(name):
    prompt = "Extract these attributes from the given string: movie name, year, resolution (e.g. 720p or 1080p). Reformat the given string by capitalising the movie names, removing any extra characters like hyphens or underscores, and following this format strictly: Movie Name (Year) [Resolution]. Here is an example: input string: ant-man_2015_1080p, output string: Ant Man (2015) [1080p]. Follow this output format exactly, do not print anything else. Input string: " + name
    response = openai.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role" : "user", "content" : prompt}]
    )

    return response.choices[0].message.content.strip()
```

```python
def rename_files(entry, new_name):
    full_entry_path = os.path.join(path, entry)
    if os.path.isfile(full_entry_path):
        os.rename(full_entry_path, os.path.join(path, f"{new_name}.torrent"))   # Change extension to '.srt' for subtitle files.
```

```python
for i in all_entries:
    final_file_name = reformat_names(i)
    rename_files(i, final_file_name)
    df = pd.concat([df, pd.DataFrame({'Old Name': [i], 'New Name': [final_file_name]})], ignore_index=True)
```
