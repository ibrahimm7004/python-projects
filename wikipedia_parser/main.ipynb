```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://en.wikipedia.org/wiki/List_of_newspapers_in_the_United_States#:~:text=Post%20(1877)-,United%20States%20newspapers%20by%20state%20and%20territory,-%5Bedit%5D"
```

```python
response = requests.get(url)    # returns a status code, 200 means 'OK'
```

```python
response.text
```

```python
def parser(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find names for all states:
    state_names = soup.find('div', class_='div-col').find_all('a')

    return state_names
```

```python
data = parser(url)
```

```python
state_names = []
for line in data:
    state_names.append(line.text)
```

```python
print(state_names)
```

```python
state_links = []
for line in data:
    state_links.append(line['href'])
```

```python
print(state_links)
```

```python
def parser(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    lines = soup.find('div', class_='div-col').find_all('a')

    states_dict = {}
    initial_url = 'https://en.wikipedia.org'

    for line in lines:
        state_name = line.text
        state_url = initial_url + line['href']
        states_dict[state_name] = state_url

    return states_dict
```

```python
final_dict = parser(url)
print(final_dict)
```
